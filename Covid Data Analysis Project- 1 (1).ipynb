```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%d-%m-%y')
```

```python
covid_df['Date'] = covid_df['Date'].dt.strftime('%d-%m-%Y')
```

```python
covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured']) + (covid_df['Deaths'])# creating new coloum and finding the Active cases
covid_df.tail(10)
```

```python
statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = "max")
```

```python
 statewise["RecoveryDate"] = statewise["Cured"]*100/statewise["Confirmed"] # creating recovery date 
```

```python
 statewise["MortalityRate"] = statewise["Deaths"]*100/statewise["Confirmed"] # creating mortality rae
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False) #sorting values
```

```python
statewise.plot() # plotting the statewise in a graph 
```

```python
vaccine = vaccine_df[vaccine_df.State != "India"]
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaccinated" : "Total"}, inplace = True)
vaccine.head()
```

```python
max_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total", ascending = False)[:10]  
max_vac
```

```python
fig = plt.figure(figsize = (15, 9))
plt.title("top 10 vacinnated states", size = 10)
axis = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Vaccinations")
plt.show()
```

```python
max_vac = vaccine.groupby("State")["Total"].sum().to_frame("Total")
max_vac = max_vac.sort_values("Total", ascending = True)[:10]  
max_vac
```

```python
fig = plt.figure(figsize = (30, 10))
plt.title("top 10 Least vacinnated states", size = 10)
axis = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = "orange")
plt.xlabel("States")
plt.ylabel("Vaccinations")
plt.show()
```
